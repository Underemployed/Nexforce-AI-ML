```python
mean_height = df['Height(Inches)'].mean()
mean_weight = df['Weight(Pounds)'].mean()

median_height = df['Height(Inches)'].median()
median_weight = df['Weight(Pounds)'].median()

mode_height = df['Height(Inches)'].mode()[0]
mode_weight = df['Weight(Pounds)'].mode()[0]

print("Mean Height: ", mean_height)
print("Median Height: ", median_height)
print("Mode Height: ", mode_height)
print(30*"-")
print("Mean Weight: ", mean_weight)
print("Median Weight: ", median_weight)
print("Mode Weight: ", mode_weight)

```
